```python
round(hotels['adr'].mean(),2)#среднее значение adr для всех бронирований в этом наборе данных
```

```python
hotels['total_stay_days'] = hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights']
```

```python
round(hotels['total_stay_days'].mean(),2)#среднее (mean) количество ночей
```

```python
hotels['total_spend'] = hotels['adr'] * hotels['total_stay_days']
#средняя общая сумма бронирования в этом наборе данных
```

```python
round(hotels['total_spend'].mean(),2)#средняя общая сумма бронирования в этом наборе данных
```

```python
hotels[hotels['total_of_special_requests'] == 5][['name','email']]
# имена и адреса email тех, кто сделал ровно 5 специальных запросов ("Special Requests")
```
